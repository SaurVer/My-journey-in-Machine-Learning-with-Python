```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('uscrime.txt',delim_whitespace=True,header='infer')
dataset
```

```python
sns.scatterplot(x='Crime',y='Time',data=dataset,hue='So')
```

```python
from scipy import stats
```

```python
z=np.abs(stats.zscore(dataset))
```

```python
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})
```

```python
zscore=pd.DataFrame(z)
#zscore
lst=[]
for i in range(0,16):
    
    lst.append(zscore[zscore[i]>2].index.values.astype(int))
lst
```

```python
sns.boxplot(x="So",y="Crime",data=dataset)
plt.title('Outliers visualisation')
```

```python
tips = sns.load_dataset("tips")

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
```
