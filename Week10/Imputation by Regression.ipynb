```python
newdata=data.dropna()
newdata.shape
```

```python
x=newdata.iloc[:,[data.columns.get_loc('Uniformity of Cell Shape'),data.columns.get_loc('Uniformity of Cell Size')]].values
y=newdata.iloc[:,data.columns.get_loc('Bare Nuclei')].values
```

```python
pd.isnull(data['Bare Nuclei'][[23, 40, 139, 145, 158, 164, 235, 249, 275, 292, 294, 297, 315, 321,
            411, 617]])
```

```python
for index in data[data['Bare Nuclei'].isnull()].index.tolist():
            #linear regression equation
            a = 0.59 * data['Uniformity of Cell Shape'][index] + 0.296 * data['Uniformity of Cell Size'][index] + .71
            # replacing "nan" values in dataframe using .set_value
            a=a.astype(int)
            data.iloc[[index],data.columns.get_loc('Bare Nuclei')]=a
```
