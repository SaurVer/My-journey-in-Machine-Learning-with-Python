```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('breast-cancer-wisconsin.data',header=None)
```

```python
data.columns=['Sample code number','Clump Thickness','Uniformity of Cell Size' ,'Uniformity of Cell Shape','Marginal Adhesion',
 'Single Epithelial Cell Size',
 'Bare Nuclei',
 'Bland Chromatin',
 'Normal Nucleoli',
 'Mitoses',
 'Class']
```

```python
data[data['Bare Nuclei']=='?'].index
```

```python
data=data.apply(pd.to_numeric,errors='coerce')
```

```python
data.iloc[[23, 40, 139, 145, 158, 164, 235, 249, 275, 292, 294, 297, 315, 321,
            411, 617],:]
```

```python
from sklearn.preprocessing import Imputer
```

```python
imputer=Imputer(missing_values='NaN', strategy='mean',axis=0)
```

```python
imputer=imputer.fit(data.iloc[:,data.columns.get_loc('Bare Nuclei'):data.columns.get_loc('Bland Chromatin')])
```

```python
data.iloc[:,data.columns.get_loc('Bare Nuclei'):data.columns.get_loc('Bland Chromatin')]=imputer.transform(data.iloc[:,data.columns.get_loc('Bare Nuclei'):data.columns.get_loc('Bland Chromatin')])
```
