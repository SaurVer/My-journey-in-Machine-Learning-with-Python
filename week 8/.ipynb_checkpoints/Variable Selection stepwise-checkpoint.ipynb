```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
import statsmodels.api as sm
```

```python
data=pd.read_csv('uscrime.txt',delim_whitespace=True)
```

```python
data.head()
```

```python
# with all the variables-Processing Part
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values

onehotencode=OneHotEncoder(categorical_features=[1])
x=onehotencode.fit_transform(x).toarray()
sc_x=StandardScaler()
x=sc_x.fit_transform(x)
```

```python

regressor=LinearRegression()
regressor.fit(x,y)
y_pred=regressor.predict(x)
rmse = np.sqrt((y - y_pred)**2).mean()
rmse
```

# I will now try Forward Selection on the basis of rmse. This way it is becoming more hit and trial method.
