```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
```

```python
data=pd.read_csv('uscrime.txt',delim_whitespace=True)
```

# Do one hot encoding on the x dataset before splitting into train test

```python
from sklearn.preprocessing import OneHotEncoder
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
onehot=OneHotEncoder(categorical_features=[1])
x=onehot.fit_transform(x).toarray()
x.shape
```

```python
from sklearn.decomposition import  PCA
pca=PCA(n_components=5)
```

```python
principal_components=pca.fit_transform(x)
principalDf=pd.DataFrame(principal_components,columns=['PC1','PC2','PC3','PC4','PC5'])
```

```python
principalDf=pd.concat([principalDf,data['Crime']],axis=1)
principalDf
```

```python
X=principalDf.iloc[:,:-1].values
Y=principalDf.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x=StandardScaler()
```

```python
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
```

```python
#rmse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)
#rmse
```

```python
r2
```

```python
print('Slope:' ,regressor.coef_)
print('Intercept:', regressor.intercept_)
```
