```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('iris.txt',delim_whitespace=True)
dataset
```

```python
x=dataset.iloc[:,[1,2,3,4]].values
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler.fit(x)
x=scaler.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
```

# Deciding the number of cluster using the elbow method

```python
wcss=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,n_init=10,max_iter=300)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
   
```

```python
sns.lineplot(range(1,10),wcss)
plt.title('Elbow Method')
plt.xlabel('No of clusters')
plt.ylabel('WCSS Value')
```

# Major kink starts at k=3

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
kmeans.fit_predict(x)
```
