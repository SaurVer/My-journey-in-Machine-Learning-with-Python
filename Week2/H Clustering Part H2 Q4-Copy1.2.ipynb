```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('iris.txt',delim_whitespace=True)
dataset
```

```python
x=dataset.iloc[:,[1,2,3,4]].values
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler.fit(x)
x=scaler.fit_transform(x)
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('dendogram')
plt.xlabel('flower')
plt.ylabel('distance')
```

# 2 seems to be the optimal number of cluster

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
HC=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
```

```python
y_HC=HC.fit_predict(x)
```

```python
y_HC
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.scatterplot(x[y_HC==0,3],x[y_HC==0,1],color='red',label='cluster1')
sns.scatterplot(x[y_HC==1,3],x[y_HC==1,1],color='blue',label='cluster2')
sns.scatterplot(x[y_HC==2,3],x[y_HC==2,1],color='green',label='cluster3')
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.scatterplot(x[y_HC==0,2],x[y_HC==0,1],color='red',label='cluster1')
sns.scatterplot(x[y_HC==1,2],x[y_HC==1,1],color='blue',label='cluster2')
sns.scatterplot(x[y_HC==2,2],x[y_HC==2,1],color='green',label='cluster3')
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.scatterplot(x[y_HC==0,2],x[y_HC==0,3],color='red',label='cluster1')
sns.scatterplot(x[y_HC==1,2],x[y_HC==1,3],color='blue',label='cluster2')
sns.scatterplot(x[y_HC==2,2],x[y_HC==2,3],color='green',label='cluster3')
```
